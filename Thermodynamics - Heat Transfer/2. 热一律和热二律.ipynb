### 热一律

>   本质是能量守恒

把热一律应用于系统的能量变化则可以表述为：**进入系统的能量减去离开系统的能量等于系统中储存能量的增量**

#### 闭口系统能量方程

公式表述 $ q = \Delta u + w $

对于简单可压缩系统而言 $ \delta q = du + pdv $

将以上式子应用于循环，得 $ \displaystyle\oint \delta u = 0 $，所以$ \displaystyle\oint \delta q = \displaystyle\oint \delta w $，$ q_{net} = w_{net} $

&nbsp;

#### 稳定流动能量方程

>   开口系统得最简单状态：流入与流出质量相等

公式表述 $ Q = \Delta H + \dfrac{1}{2} m \Delta c^2 + mg\Delta z + W_s $

微量形式 $ q = \Delta h + \dfrac{1}{2} \Delta c^2 + g\Delta z + w_s $，考虑到$ \Delta h = \Delta u + \Delta (pv) $

方程被重新表现为$ q - \Delta u = \dfrac{1}{2} \Delta c^2 + g\Delta z + w_s + \Delta (pv) $

其中，$ w_t = \dfrac{1}{2} \Delta c^2 + g\Delta z + w_s $是技术上可以利用的功为**技术功**。

则又可以将原式子简写为$ w - w_t = \Delta (pv) $，因而$ w_t  = \displaystyle\int_1^2 pdv + p_1v_1 - p_2v_2 = \displaystyle\int_1^2 pdv - \displaystyle\int_1^2 d(pv) = -\displaystyle\int_1^2 vdp $

引入技术功得概念之后，方程可以写为$ q = h_2 - h_1 + w_t $，对于微元过程可写为$ \delta q = dh - vdp $

##### 应用

$ q = \Delta h + \dfrac{1}{2} \Delta c^2 + g\Delta z + w_s $

-   汽轮机

一类动力机构，过程中工质基本不会换热$ q \approx 0 $，且忽略机械能变化

则$ w_s = h_1 - h_2 $ 工质所做的轴功等于焓降

-   压力机

如泵与风机，过程中外界对工质做功加压，考虑为绝热过程，且忽略机械能的变化

则$ -w_s = h_2 - h_1 $ 工质绝热压缩时，所消耗的轴功等于焓的增加

-   换热器

水流过散热器时，轴功不会进行交换，且忽略机械能

则$ q = h_1 - h_2 $ 水在散热器中的散热量的焓差

#### 热一律相关思考题

1.  工质膨胀是否一定对外做功？ 

不一定。对外做功应该有两个条件：一是气体膨胀；二是有做功的对象（若是真空自由膨胀那么并不会做功）

1.  定容过程一定不做功吗？

不一定。对于气体而言，根据$ w =\int pdv $可得不会做功；但是对于液体，如水轮机的情况而言，容积不变，但是仍然会做功

1.  定温过程是否一定不传热？

不一定。有些气化液化过程是等温过程，在这些过程中会有热量传递的发生。

### 热二律

>   本质是热现象各种过程进行的方向、条件与进行的限度

<u>自发的传热过程是不可逆的</u>

-   克劳修斯**表述**：热量不可能自发的、不花任何代价的从低温物体转向高温物体
-   开尔文**表述**：不可能从单一热源吸热，使之全部转化为功而不留下任何变化的热力循环永动机

否定了**第二类永动机**：利用大气、海洋作为单一热源，从中汲取能量转变为功（不违背热一律、违背热二律）

#### 卡诺循环

>   由两个可逆等温过程和可逆绝热过程组成

用$ \eta_c $作为卡诺循环的效率，由循环的T-s较容易得到热效率

公式即为$ \eta_c = \dfrac{w_{net}}{q_1} = 1- \dfrac{q_2}{q_1} = 1- \dfrac{T_2 \Delta s_{ab}}{T_1 \Delta s_{ab}} = 1 - \dfrac{T_2}{T_1} $

重要结论：

1.  卡诺循环效率取决于高低热源的温度，提高冷源温度（或降低热源温度，也可以所理解为增大温差），可以提高效率
2.  根据等式可得$ T_2 < T_1 $，循环效率必定小于1，不可能将热完全转化功
3.  不可呢从一个单独的热源吸热并完全转化功

卡诺循环作为一种可逆循环必定会大于其他一切不可逆热机的效率

### 熵方程和孤立系统熵增原理

#### 熵的导出

卡诺循环，可逆热机从高温热源T1吸热Q1，向低温热源T2放热Q2

则热效率为$ \dfrac{Q_2}{Q_1} = \dfrac{T_2}{T_1} $，由绝对值转变为根据吸放热确定符号则原式等于$ \dfrac{Q_1}{T_1} + \dfrac{Q_2}{T_2} = 0 $

微分化之后$ dS = \dfrac{\delta Q_{rev}}{T_r} =  \dfrac{\delta Q_{rev}}{T} $，在循环之中，$ \displaystyle\oint dS = 0 $

**可逆过程**中，熵的变化$ \Delta s = s_2 - s_1 = \displaystyle\int_1^2 \dfrac{\delta q_{rev}}{T_r} $，与过程路径无关

**不可逆循环**中，有些微循环不可逆，则$ \displaystyle\oint dS < 0 $

&nbsp;

#### 熵流与熵产

不可逆的情况下，熵变量大于不可逆过程中工质与热源交换能量与温度比值的积分

即为$ \Delta s = s_2 - s_1 \geq \displaystyle\int_1^2 \dfrac{\delta q}{T_r} = s_f + s_g$

-   $ s_f = \displaystyle\int_1^2 \dfrac{\delta q}{T_r} $，flow熵流，系统与外界换热对系统的贡献
-   $ s_g $，generation熵产，不可逆性对于熵变的贡献，非负的，是系统**不可逆性**的一种度量

&nbsp;

&nbsp;

#### 熵方程和熵增原理

##### 开口系统

熵方程公式$ dS_{CV} = s_1 \delta m_1 - s_2 \delta m_2 + \delta S_f + \delta S_g $

**系统熵的平衡**即为流入系统的熵减去流出系统的熵、热流造成的热熵流以及不可逆造成的熵产的总和

工程上**稳态稳流系统**，整体系统熵变$ dS_{CV} = 0 $且流入与流出质量相等$  \delta m_1 =  \delta m_2 $

在一般情况下，熵产较不容易计算则$ S_g = m(s_2 - s_1)  - S_f$

若再增加一个条件**整个稳流系统绝热**，则热熵流为0，所以可得$ s_2 - s_1 = s_g \geq 0 $

在不可逆的情况下，熵产大于0，则出口截面的比熵大于入口截面比熵

&nbsp;

##### 闭口系统

与外界无物质交换，则$  \delta m_1 =  \delta m_2 = 0 $，则$ ds = \delta S_f + \delta S_g $

所以分前后两种状态（以1，2表示）则$ \Delta S_{iso} = \Delta S_{1-2} = S_2 - S_1 = S_f + S_g \geq 0 $

>   **熵增原理**：孤立系统内的熵只可能增大，不可能减小，极限状况是孤立系统内一切过程全部可逆，系统熵保持不变！
