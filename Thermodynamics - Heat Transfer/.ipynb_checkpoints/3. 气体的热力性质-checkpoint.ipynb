### 理想气体的比热容

#### 比热容

-   公式直接$ c = \dfrac{\delta q}{d T} $，1kg物质升高1度所需要吸收的能量，单位是$ J/(kg \cdot K) $
-   1mol物质的热容为**摩尔热容**$ C_m $，单位是$ J/(mol \cdot K) $
-   标准状态下，1立方米的热容是**体积热容**$ C' $，单位是$ J/(m^3 \cdot K) $
-   三种热容的关系$ C_m = Mc = V_{m,0}C' $，其中$ V_{m,0} = 22.4 \times 10^{-3} m^3/mol $

&nbsp;

#### 理想气体的比热容 & 迈耶公式

##### 比定压/容比热容

>   在热力设备中，等压和等容两种状态较为常见，则给出两种定义比定压/容比热容

-   比定压比热容

$ c_p = \dfrac{\delta q_p}{dT} $，联合热力学第一定律变式（引入**技术功**）：$ \delta q = dh - vdp $

则可得$ c_p = \dfrac{\delta q_p}{dT} = (\dfrac{\alpha h}{\alpha T})_p $

-   比定容比热容

$ c_V = \dfrac{\delta q_v}{dT} $，联合热力学第一定律：$ \delta q = du + pdv $

则可得$ c_V = \dfrac{\delta q_v}{dT} = (\dfrac{\alpha u}{\alpha T})_v$

&nbsp;

##### 迈耶公式

-   在理想气体状态下，上述两式均可以化为微分模式 $ c_p = \dfrac{dh}{dT} $ 和 $ c_V = \dfrac{du}{dT} $
-   考虑1kg理想气体从温度T1分别经过等压和等容升高1K

则 $ q_p = c_p $ 和 $ q_v = c_v $

所以$ q_p - q_V = c_p - c_V = (h_2 - h_1) - (u_2 - u_1 + w) = R_g = c_p - c_V $

在实际过程中，定体积比热容难测，则$ C_{p,m} - C_{V,m} = MR_g = R $

&nbsp;

##### 比热容比

表示符号为$ \gamma $，在理想气体中，比热容比等于绝热系数$ \kappa $，则$ \gamma = \kappa = \dfrac{c_p}{c_V} $

所以可以导出公式 $ c_V = \dfrac{R_g}{\gamma - 1} $ 和 $ c_p = \gamma c_V = \dfrac{\kappa R_g}{\gamma - 1} $

##### 相关常用公式

$ R_g = \dfrac{R}{M} $，比热容比$ \gamma = \dfrac{i+2}{i} $，单原子分子是5/3，双原子7/5，多原子9/7

### 水蒸气

#### 饱和状态、温度、压力

-   当汽化和凝结处于动态平衡，则是**饱和状态**，对应液体叫做**饱和液体**，对应气体叫做**干饱和蒸气**。
-   与**饱和温度**对应的蒸气对液面的压力叫做**饱和压力**，属于一一对应关系。
-   **<u>对比</u>**：对于理想气体而言，需要有两个参数，才能真正确定一个物理状态。

&nbsp;

#### 汽化潜热

##### 相变全过程

1.  $ t < t_s $，未饱和液：低于所处压力下的饱和温度
2.  $ t = t_s $，饱和液：处于所处压力下的饱和温度
3.  $ t = t_s $，湿饱和蒸汽：饱和液和干饱和蒸汽的混合物
4.  $ t = t_s $，干饱和蒸汽：处于饱和状态的蒸汽 
5.  $ t > t_s $，过热蒸汽：温度高于饱和温度的蒸汽，$ t - t_s = d $ 为过热度

-   1kg**饱和液体**全部转变为同温下的**干饱和蒸汽**所吸收的热量后~续图见书本`P76-77`
