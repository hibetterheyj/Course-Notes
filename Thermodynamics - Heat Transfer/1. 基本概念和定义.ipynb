### 状态和状态参量

#### 热力系统

-   热力学中研究的对象：**热力系统（系统、体系）** 属于人为确定的！
-   与系统发生质量/能量交换的周围物质系统：**外界**
-   与系统交换能量的物质系统：**外界热源**
-   携带热能的工作物质：**工质** 
-   *e.g.* 气液过渡转化的<u>蒸气</u>、燃机<u>燃气</u>、制冷装置的<u>氨蒸气</u> 物质三态中**气态、水、卤代烷烃（制冷剂）**最适宜！*要求*：**膨胀性**（将简单无规则的原子运动转为宏观运动）、**流动性**……
-   **热力系统分类**

1.按<u>系统与外界质量交换</u>分：**闭口系**（没有质量越过边界，控制**质量**交换cm, *control mass*）VS  **开口系**（通过边界且与外界有质量交换,控制体积不变cv）

2.按<u>系统与外界能量交换</u>分：**绝热系**（与外界无热量交换）VS **孤立系**（与外界无任何质能交换，宇宙）

3.按<u>组成物质</u>分：**简单可压缩系**（系统内无化学反应，且与外界的功交换仅为体积变化功，即只有通过传热做功）

-   注意：

    闭口系与系统内质量不变（关键在于**工质**二字，闭口系工质不变，内质量则是总质量不变）

    开口系与绝热系的关系（可以有质量交换而没有热量交换）

#### 平衡状态

>   没有外界作用下工质的宏观性质保持不变的状态——这个系统同时出于**热平衡**和**力平衡**的状态。如果受到外界作用，<u>系统的内压力和温度不均等将会破坏平衡系统</u>。
>
>   特殊概念：**局部平衡**

**注意点**：

1.  系统平衡但是工质不一定均匀分布：如水与水蒸气在临界温度的转化
2.  系统稳定不一定平衡：如一根两头温度分别是50和100度的钢杆，整体温度，但因为有外部热源不处于平衡状态

#### 状态参数

对于出于平衡状态的工质可以用一组宏观的物理量进行描述，它们即是状态参数。常用的有压强p，温度T，体积V，热力学能U，焓H和熵S（都属于**宏观量**！）大写一般表示与**质量**有关~

-   **与达到状态的过程无关，闭合线积分为0**
-   平衡状态下的**简单可压缩系统**只有两个独立状态参量。
-   状态参量可以分为两类：**强度量**和**广延量**（分别与物质的量**无/有**关）——广延量的比性质具有强度量性质
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;1.强度量：压力**p**，热力学温度**T**，比体积$ \nu = V/M $
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;2.广延量：体积**V**，热力学能**U**，焓**H**和熵**S**

### 基本状态参数

#### 1.温度

-   热力学温度，符号是T，单位是K。
-   热力学温标以水的三相点作为基本定点，定义温度为273.16K
-   摄氏度温标并用，符号是t，单位是度，存在着一定的关系：$ t = T - T_0 $，其中$ T_0 = 273.15 K $
-   水的三相点是0.01度，也是273.16K

#### 2.压力

-   主要的单位是Pa，其中$ 1Pa = 1N/m^2 $
-   还可能用到如$ 1bar = 10^5Pa $，$ 1atm = 101325Pa \approx 1 bar  = 760mmHg = 10.34m$水柱
-   需要区分的概念：绝对压力、表压力以及真空度

#### 3.比体积

-   1kg工质所具有的体积，符号是$ v $，单位是$ m^3/kg $，按定义可以得$ v = \dfrac{V}{m} $
-   对应的一个概念：**密度**！其定义式$ \rho = \dfrac{m}{V} = \dfrac{1}{v} $

### 状态方程

-   理想气体状态方程 $ pv = R_g T $，p压强，v比体积，Rg体积常数
-   公式也可以写为$ pV = mR_g T $ 或 $ pV = nRT $，其中$ R = M \cdot R_g = 8.3145 J/(mol \cdot K) $

### 热力学能、焓与熵

#### 热力学能

-   1kg气体的热力学能为$ u $，质量为m的气体热力学能$ U = Mu $，工程中较为关注的是$ \Delta U $
-   总能公式$ E = U + E_k + E_P = U + \dfrac{1}{2}mc^2 +mgz $，以单位质量来写$ e = u + \dfrac{1}{2}c^2 +gz $

&nbsp;

#### 焓

##### 1. 推动功和流动功

-   推动功：工质向运动方向上前进需要承受来自其他工质的阻碍力，如进口的$ p_1 V_1 $，出口$ p_2 V_2 $
-   流动功：开口系保持工质流进、流出必须付出的代价，为两者之差$ p_2 V_2 - p_1 V_1 $

##### 2.焓

直接定义公式$ H = U + pV $，单位是J

-   对于理想气体而言，1kg工质的焓叫做比焓$ h = \dfrac{H}{m} = u + pv = u + R_g T $
-   焓代表着工质流入（或流出）开口系时传入（或传出）系统的能量
-   无法测定准确值，则在工程中比较关注的时$ \Delta H $

##### 3.熵

克劳修斯直接定义公式$ dS = \dfrac{\delta Q_{rev}}{T} $，单位是J/K

-   对于可逆系统而言，$ Q = \int mTds $
-   绝热根据方程可以看作是一个**定熵**过程

### 可逆过程

-   准静态过程：偏离平衡状态无穷小，随时恢复平衡的状态变化过程
-   可逆过程：可以看作是准静态过程加上没有**耗散**——系统经历一个过程后，系统可以沿原过程反向进行而不在外界留下任何影响。
-   耗散效应：在准静态的膨胀过程中，工质膨胀功的一部分克服摩擦力而耗散转变成热！

&nbsp;

&nbsp;

### 功和能量

-   系统向外做功为正、外界对系统做功为负
-   1kg气体在整个膨胀过程之中所做的功为$ w = \displaystyle\int pdv $
-   1kg气体在一个过程中放热量$ q = \int_1^2 Tds $

### 循环

#### 分类

##### 正向循环

>   目的是输出功，又称动力循环。

膨胀做功大于外界做功，工质吸热大于放热

对应效率$ \eta_1 = \dfrac{w_{net}}{q_1} $，即循环净做功与循环吸热量的比，效率恒小于1

##### 逆向循环

>   目的是把热量从低温传向高温系统，分为制冷循环或热泵循环。

膨胀做功小于外界做功，工质吸热小于放热

制冷效率$ \varepsilon = \dfrac{q_c}{w_{net}} $，即自低温系统吸热量与循环净功的比，效率可大可小

制冷效率$ \varepsilon ' = \dfrac{q_1}{w_{net}} $，即向高温系统的放热量量与循环净功的比，效率恒大于1
